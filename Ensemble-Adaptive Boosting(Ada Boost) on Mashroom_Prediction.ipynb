```python
# Data Preprocessing : 

# Data Cleaning :

# In this dataset wrong data, wrong data types are not there.

print(df.isnull().sum())         # No any null value is there
print("-"*100)
print(df.duplicated().sum())     # No any duplicates value is there
print("-"*100)

# Data Wrangling : 

# In this dataset we have all columns with discrete categorical values, so we have to do the encoding, we will use here dummy encoding
# Here more input columns are there for encoding comparatevely to the Penguine_Dataset, So we should not do the enconding with taking direct input columns names in get_dummies() function,
# we should do encoding by dropping output column directly, that will gives us all input columns and also takes comparitively less time.

df1 = pd.get_dummies(data = df.drop("class",axis=1),drop_first=True)
print(df1)
```

```python
# model(machine) can understand True & False from the encoding as the 1's and 0's internally, 
# So here is no need of converting True and False into the 1's and 0's resp.


X = df1    # or  pd.get_dummies(data = df.drop("class",axis=1),drop_first=True)
print(X)
print("-"*100)

y = df["class"]
print(y)
print("-"*100)


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=101)


# Modelling :  for default model  

from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(X_train,y_train)


# Prediction :  for default model 

ypred_train = model.predict(X_train)
ypred_test = model.predict(X_test)


# Evaluation :  for default model

train_accuracy = model.score(X_train,y_train)
print("train_accuracy:",train_accuracy)
test_accuracy = model.score(X_test,y_test)
print("test_accuracy:",test_accuracy)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv=5)
print("cross_val_score:",scores.mean())

```

```python
# We have to identify the best Hypeparameters and important columns from datset for best AdaBoost model 

from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import AdaBoostClassifier
estimator = AdaBoostClassifier()

grid_param = {"n_estimators":list(range(1,25))}
ad_classifier = GridSearchCV(estimator,grid_param,cv=5,scoring="accuracy")

ad_classifier.fit(X_train,y_train)
ad_classifier.best_params_

```

```python
ad_classifier.best_estimator_.feature_importances_
```

```python
feats = pd.DataFrame(data = ad_classifier.best_estimator_.feature_importances_,
                    index = X.columns,
                    columns = ["Important Features"])
print(feats)
print("-"*100)

imp_feats = feats[feats["Important Features"]>0]
print(imp_feats)

imp_feats_list = imp_feats.index.tolist()      #  or list(imp_feats)  it will also give same output(answer)
```

```python
X_new = X[imp_feats_list]
print(X_new)
print("-"*100)

y = df["class"]   # y will be as it is (same as old y)


# For best final model we have to do again train_test_split, modelling, prediction, evaluation with the identified best Hyperparameter and important features

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=6)


# Modelling : with best Hyperparameter and Important Features 

from sklearn.ensemble import AdaBoostClassifier
final_model = AdaBoostClassifier()
final_model.fit(X_train,y_train)


# Prediction :  with best Hyperparameter and Important Features

ypred_train = final_model.predict(X_train)
ypred_test = final_model.predict(X_test)


# Evaluation :  with best Hyperparameter and Important Features

train_accuracy = final_model.score(X_train,y_train)
print("train_accuracy:",train_accuracy)
test_accuracy = final_model.score(X_test,y_test)
print("test_accuracy:",test_accuracy)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(final_model,X,y,cv=5)
print("cross_val_score:",scores.mean())
```
